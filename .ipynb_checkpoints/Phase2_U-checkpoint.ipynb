## U0. Scope and Deliverables (Phase 2)
- Build a reproducible data-understanding workflow for movie recommendation.
- Audit data quality (missing values, duplicates, schema consistency, time coverage).
- Quantify recommendation-specific challenges (sparsity, long-tail, cold-start, drift).
- Produce concrete decisions for the next phase (cutoffs + chronological split strategy).

**Expected input**: MovieLens ratings and movies tables.
- `data/ml-25m/ratings.csv`
- `data/ml-25m/movies.csv`

If files are not present, this notebook falls back to synthetic data so all steps can still be executed.


```python

from pathlib import Path
import re

import numpy as np
import pandas as pd
import matplotlib.pyplot as plt

pd.set_option("display.max_columns", 50)
pd.set_option("display.width", 140)
plt.style.use("ggplot")

RNG = np.random.default_rng(42)
DATA_ROOT = Path("data/ml-25m")
RATINGS_FILE = DATA_ROOT / "ratings.csv"
MOVIES_FILE = DATA_ROOT / "movies.csv"
```

```python

def _build_synthetic_data(
    n_users: int = 1500,
    n_items: int = 1200,
    min_user_events: int = 8,
    max_user_events: int = 70,
):
    """Generate a realistic sparse interaction table with long-tail behavior."""

    item_ids = np.arange(1, n_items + 1)
    item_weights = np.power(np.arange(1, n_items + 1), -1.08)
    item_weights = item_weights / item_weights.sum()

    rows = []
    ts_start = int(pd.Timestamp("2021-01-01").timestamp())
    ts_end = int(pd.Timestamp("2025-01-01").timestamp())

    for user_id in range(1, n_users + 1):
        n_events = int(RNG.integers(min_user_events, max_user_events + 1))
        sampled_items = RNG.choice(item_ids, size=n_events, replace=True, p=item_weights)
        sampled_ratings = RNG.choice(
            [0.5, 1, 1.5, 2, 2.5, 3, 3.5, 4, 4.5, 5],
            size=n_events,
            p=[0.01, 0.02, 0.03, 0.05, 0.08, 0.18, 0.22, 0.20, 0.14, 0.07],
        )
        sampled_ts = RNG.integers(ts_start, ts_end, size=n_events)

        for movie_id, rating, timestamp in zip(sampled_items, sampled_ratings, sampled_ts):
            rows.append((user_id, int(movie_id), float(rating), int(timestamp)))

    ratings = pd.DataFrame(rows, columns=["user_id", "item_id", "rating", "timestamp"])

    genres_pool = [
        "Action", "Adventure", "Animation", "Comedy", "Crime", "Drama",
        "Fantasy", "Horror", "Mystery", "Romance", "Sci-Fi", "Thriller",
    ]

    movie_rows = []
    for movie_id in item_ids:
        n_genres = int(RNG.integers(1, 4))
        selected = RNG.choice(genres_pool, size=n_genres, replace=False)
        year = int(RNG.integers(1970, 2025))
        movie_rows.append((movie_id, f"Synthetic Movie {movie_id} ({year})", "|".join(selected)))

    movies = pd.DataFrame(movie_rows, columns=["item_id", "title", "genres"])
    source = "synthetic"
    return ratings, movies, source


def load_ratings_and_movies(ratings_path: Path, movies_path: Path):
    if ratings_path.exists() and movies_path.exists():
        ratings = pd.read_csv(ratings_path)
        movies = pd.read_csv(movies_path)

        ratings = ratings.rename(columns={"userId": "user_id", "movieId": "item_id"})
        movies = movies.rename(columns={"movieId": "item_id"})

        required_rating_cols = {"user_id", "item_id", "rating", "timestamp"}
        required_movie_cols = {"item_id", "title", "genres"}

        missing_rating = required_rating_cols - set(ratings.columns)
        missing_movie = required_movie_cols - set(movies.columns)

        if missing_rating:
            raise ValueError(f"ratings.csv missing columns: {sorted(missing_rating)}")
        if missing_movie:
            raise ValueError(f"movies.csv missing columns: {sorted(missing_movie)}")

        ratings = ratings[["user_id", "item_id", "rating", "timestamp"]].copy()
        movies = movies[["item_id", "title", "genres"]].copy()

        source = "movielens"
        return ratings, movies, source

    return _build_synthetic_data()


ratings, movies, data_source = load_ratings_and_movies(RATINGS_FILE, MOVIES_FILE)

ratings["user_id"] = ratings["user_id"].astype("int64")
ratings["item_id"] = ratings["item_id"].astype("int64")
ratings["rating"] = ratings["rating"].astype("float64")
ratings["timestamp"] = ratings["timestamp"].astype("int64")

movies["item_id"] = movies["item_id"].astype("int64")
movies["title"] = movies["title"].astype("string")
movies["genres"] = movies["genres"].fillna("(no genres listed)").astype("string")

print(f"Data source: {data_source}")
print(f"Ratings shape: {ratings.shape}")
print(f"Movies shape: {movies.shape}")
ratings.head()
```


## U1. Schema and Data Quality Audit
We validate structure before any model-related decisions.


```python

quality = pd.DataFrame({
    "column": ratings.columns,
    "dtype": [str(ratings[c].dtype) for c in ratings.columns],
    "missing": [int(ratings[c].isna().sum()) for c in ratings.columns],
    "missing_pct": [float(ratings[c].isna().mean() * 100) for c in ratings.columns],
})

duplicate_rows = int(ratings.duplicated().sum())
rating_range_ok = bool(((ratings["rating"] >= 0.5) & (ratings["rating"] <= 5.0)).all())

print("Duplicate rows in ratings:", duplicate_rows)
print("Ratings within [0.5, 5.0]:", rating_range_ok)
quality
```

```python

ratings["event_time"] = pd.to_datetime(ratings["timestamp"], unit="s", errors="coerce", utc=True)

time_min = ratings["event_time"].min()
time_max = ratings["event_time"].max()

print("Time coverage:", time_min, "->", time_max)
ratings[["user_id", "item_id", "rating", "event_time"]].head()
```


## U2. Core Recommendation Diagnostics
The key feasibility indicators for collaborative filtering are user/item cardinality, matrix sparsity, and long-tail concentration.


```python

n_users = ratings["user_id"].nunique()
n_items = ratings["item_id"].nunique()
n_interactions = len(ratings)
density = n_interactions / float(n_users * n_items)

print(f"Users: {n_users:,}")
print(f"Items: {n_items:,}")
print(f"Interactions: {n_interactions:,}")
print(f"Interaction-matrix density: {density:.6f} ({density*100:.4f}%)")
```

```python

user_activity = ratings.groupby("user_id").size().rename("interactions")
item_popularity = ratings.groupby("item_id").size().rename("interactions")

fig, axes = plt.subplots(1, 2, figsize=(13, 4.2))

axes[0].hist(user_activity.values, bins=60, log=True, color="#1f77b4", edgecolor="white")
axes[0].set_title("Interactions per User")
axes[0].set_xlabel("# interactions")
axes[0].set_ylabel("Frequency (log scale)")

axes[1].hist(item_popularity.values, bins=60, log=True, color="#ff7f0e", edgecolor="white")
axes[1].set_title("Interactions per Item")
axes[1].set_xlabel("# interactions")
axes[1].set_ylabel("Frequency (log scale)")

plt.tight_layout()
plt.show()

summary = pd.DataFrame({
    "metric": [
        "median interactions/user",
        "median interactions/item",
        "p90 interactions/user",
        "p90 interactions/item",
    ],
    "value": [
        float(user_activity.median()),
        float(item_popularity.median()),
        float(user_activity.quantile(0.9)),
        float(item_popularity.quantile(0.9)),
    ],
})
summary
```

```python

# Concentration: how much traffic is captured by top-10% items
item_sorted = item_popularity.sort_values(ascending=False)
cutoff = max(1, int(len(item_sorted) * 0.10))
share_top10 = item_sorted.iloc[:cutoff].sum() / item_sorted.sum()

print(f"Top 10% items account for {share_top10:.2%} of interactions")
```


## U3. Signal Characterization
For ranking, we treat interactions as implicit feedback and inspect explicit rating behavior as an auxiliary quality signal.


```python

ratings["positive_implicit"] = (ratings["rating"] >= 4.0).astype(int)

fig, axes = plt.subplots(1, 2, figsize=(13, 4.2))

axes[0].hist(ratings["rating"].values, bins=np.arange(0.25, 5.76, 0.5), color="#2ca02c", edgecolor="white")
axes[0].set_title("Explicit Rating Distribution")
axes[0].set_xlabel("Rating")
axes[0].set_ylabel("Count")

pos_share = ratings["positive_implicit"].mean()
axes[1].bar(["non-positive", "positive (>=4.0)"], [1 - pos_share, pos_share], color=["#d62728", "#9467bd"])
axes[1].set_title("Implicit Positive Ratio")
axes[1].set_ylabel("Share")

plt.tight_layout()
plt.show()

print(f"Positive implicit share (rating >= 4.0): {pos_share:.2%}")
```


## U4. Temporal Behavior and Drift Risk
A recommendation system is sensitive to time drift. We inspect monthly interaction volume and mean rating.


```python

monthly = (
    ratings.dropna(subset=["event_time"])
    .assign(month=lambda d: d["event_time"].dt.to_period("M").dt.to_timestamp())
    .groupby("month")
    .agg(interactions=("item_id", "size"), mean_rating=("rating", "mean"))
    .reset_index()
)

fig, ax1 = plt.subplots(figsize=(13, 4.5))
ax2 = ax1.twinx()

ax1.plot(monthly["month"], monthly["interactions"], color="#1f77b4", label="interactions")
ax2.plot(monthly["month"], monthly["mean_rating"], color="#ff7f0e", label="mean_rating")

ax1.set_title("Monthly Interactions and Mean Rating")
ax1.set_xlabel("Month")
ax1.set_ylabel("Interactions", color="#1f77b4")
ax2.set_ylabel("Mean rating", color="#ff7f0e")

ax1.tick_params(axis="x", rotation=30)
plt.tight_layout()
plt.show()

monthly.tail(6)
```


## U5. Cold-Start and Metadata Coverage
We quantify low-signal entities and inspect metadata quality for content-based fallback.


```python

MIN_EVENTS = 5

cold_users_share = float((user_activity < MIN_EVENTS).mean())
cold_items_share = float((item_popularity < MIN_EVENTS).mean())

print(f"Users with <{MIN_EVENTS} interactions: {cold_users_share:.2%}")
print(f"Items with <{MIN_EVENTS} interactions: {cold_items_share:.2%}")

metadata_coverage = pd.DataFrame({
    "metric": ["missing title", "missing genres", "'(no genres listed)'"],
    "count": [
        int(movies["title"].isna().sum()),
        int(movies["genres"].isna().sum()),
        int((movies["genres"] == "(no genres listed)").sum()),
    ],
})
metadata_coverage
```

```python

# Basic item-level feature table for understanding feature relationships
item_stats = ratings.groupby("item_id").agg(
    rating_count=("rating", "size"),
    mean_rating=("rating", "mean"),
    rating_std=("rating", "std"),
).reset_index()

movies_features = movies.copy()
movies_features["release_year"] = (
    movies_features["title"].str.extract(r"\((\d{4})\)", expand=False).astype("float64")
)
movies_features["n_genres"] = movies_features["genres"].apply(
    lambda x: 0 if x == "(no genres listed)" else len(str(x).split("|"))
)

item_profile = item_stats.merge(
    movies_features[["item_id", "release_year", "n_genres"]],
    on="item_id",
    how="left",
)

corr_cols = ["rating_count", "mean_rating", "rating_std", "release_year", "n_genres"]
corr = item_profile[corr_cols].corr(numeric_only=True)

fig, ax = plt.subplots(figsize=(7.2, 5.8))
im = ax.imshow(corr.values, cmap="coolwarm", vmin=-1, vmax=1)
ax.set_xticks(range(len(corr_cols)))
ax.set_yticks(range(len(corr_cols)))
ax.set_xticklabels(corr_cols, rotation=40, ha="right")
ax.set_yticklabels(corr_cols)
ax.set_title("Item-Level Correlation Matrix")
fig.colorbar(im, ax=ax, fraction=0.046, pad=0.04)
plt.tight_layout()
plt.show()

corr.round(3)
```


## U6. Draft Decisions for Next Phase (A: Algorithm Selection)
The U phase should end with concrete and testable preprocessing/splitting decisions.


```python

MIN_USER_INTERACTIONS = 5
MIN_ITEM_INTERACTIONS = 5

user_counts = ratings.groupby("user_id").size()
item_counts = ratings.groupby("item_id").size()

eligible_users = user_counts[user_counts >= MIN_USER_INTERACTIONS].index
eligible_items = item_counts[item_counts >= MIN_ITEM_INTERACTIONS].index

filtered = ratings[
    ratings["user_id"].isin(eligible_users) & ratings["item_id"].isin(eligible_items)
].copy()

if filtered["timestamp"].notna().all():
    t80 = filtered["timestamp"].quantile(0.80)
    t90 = filtered["timestamp"].quantile(0.90)

    train = filtered[filtered["timestamp"] <= t80]
    val = filtered[(filtered["timestamp"] > t80) & (filtered["timestamp"] <= t90)]
    test = filtered[filtered["timestamp"] > t90]
else:
    # Fallback if timestamps are unavailable
    shuffled = filtered.sample(frac=1.0, random_state=42)
    n = len(shuffled)
    train = shuffled.iloc[: int(0.8 * n)]
    val = shuffled.iloc[int(0.8 * n): int(0.9 * n)]
    test = shuffled.iloc[int(0.9 * n):]

split_stats = pd.DataFrame(
    {
        "split": ["train", "validation", "test"],
        "rows": [len(train), len(val), len(test)],
        "unique_users": [train.user_id.nunique(), val.user_id.nunique(), test.user_id.nunique()],
        "unique_items": [train.item_id.nunique(), val.item_id.nunique(), test.item_id.nunique()],
    }
)

print(f"Rows before filtering: {len(ratings):,}")
print(f"Rows after filtering:  {len(filtered):,}")
split_stats
```


## U7. Reflection Snapshot (for Portfolio Text)
**What we learned in U phase**
- The interaction matrix is sparse and long-tailed, so popularity bias must be controlled in evaluation.
- Cold-start entities exist and require content-based fallback (metadata quality directly affects this).
- Temporal dynamics are non-trivial, so chronological splits are mandatory to avoid leakage.

**Implementation decisions carried to A phase**
- Use filtered training data with minimum user/item interaction cutoffs.
- Evaluate baselines and candidate models on chronological train/validation/test splits.
- Track both ranking quality (NDCG@k/HR@k/MAP) and system constraints from Q phase.
