## 2. Streamlit App Concept (User Flows and Core Screens)

The Streamlit app should present a simple entry experience while still supporting the three required recommendation modes from Q-Phase and C-Phase: a returning-user personalized flow, a new-user/cold-start flow, and a similar-movie exploration flow. The main design principle for v1 is that the app should make the user choose a clear starting intent, then map that intent to the correct backend mode without exposing technical details.

For a **returning user**, the UI should allow a user identifier (or a preloaded user selection) and optionally additional filters. The app then uses the hybrid personalized path, applies filters, and explains results using reason text such as "Because you liked X" or a fallback reason when collaborative evidence is weak.

For a **new user**, the UI should not request a user ID. Instead, it should ask for a few liked titles (optional but encouraged) and preference filters such as genre or year range. If the input is too weak for robust personalization, the app should move intentionally into the cold-start route rather than pretending the results are deeply personalized.

For a **similar-movie search**, the user should search for a title and request movies similar to that title. This mode is useful for exploration and demo clarity because it is easy to understand and does not require stored history. In v1, the similarity explanation should stay plain and honest (genre/theme similarity based on available metadata), especially because the project intentionally deferred richer content features.

The app should use a small set of consistent screens/components: an input panel (mode selection + search or user selection), a filter panel, a recommendation list area, and an explanation panel or inline explanation text on each recommendation card. This keeps the interface stable across modes while allowing the backend logic to change based on the selected route.

### Streamlit UI Wireframe (v1 Concept)

```text
+--------------------------------------------------------------------------------+
| Movie Recommendation App (Streamlit)                                           |
+--------------------------------------------------------------------------------+
| Sidebar / Left Panel                                                           |
|  [Mode selector]  (Returning User | New User | Similar Movie)                  |
|                                                                                |
|  If Returning User:                                                             |
|    [User ID selectbox / search]                                                 |
|                                                                                |
|  If New User:                                                                   |
|    [Liked movies multiselect / search]                                          |
|    [Optional genre preferences]                                                 |
|                                                                                |
|  If Similar Movie:                                                              |
|    [Movie title search box]                                                     |
|                                                                                |
|  Shared filters:                                                                |
|    [Genres] [Year range] [Min rating threshold] [Runtime if available]         |
|                                                                                |
|  [Get Recommendations] button                                                   |
+--------------------------------------------------------------------------------+
| Main Results Area                                                               |
|  Status / context message (mode + fallback note if needed)                      |
|                                                                                |
|  Recommendation cards (Top-N):                                                  |
|   - Poster (optional / placeholder)  Title (Year)  Genres                      |
|     Reason: "Because you liked ..." / "Popular in genre ..." / "Similar to..."|
|   - ...                                                                         |
|                                                                                |
|  Explanation panel (optional expandable area):                                  |
|   - Why this mode was used                                                      |
|   - Filters applied                                                             |
|   - Fallback used? (yes/no + reason)                                            |
+--------------------------------------------------------------------------------+
```

This wireframe shows the main UI structure for the Streamlit app and how the same layout supports three different entry modes. The user first clicks a mode and provides either a user ID, liked titles, or a source movie title; behind the scenes, the app uses that input to choose the correct recommender route and fallback path; on screen, the user sees a Top-N list with plain-language reasons and a short status message so the recommendations do not feel random.

### Screen and Component Responsibilities

This table shows how each visible UI component maps to a backend dependency and a user-facing output. It matters for implementation because it prevents UI work from drifting away from the recommender logic: every control should exist for a clear recommendation or explanation purpose.

## 3. Request-to-Result Inference Flow (App Runtime)

The app should treat each recommendation request as a small pipeline with explicit stages. This keeps the implementation easier to test and makes it clearer where caching, fallback logic, and error handling belong.

```text
User clicks "Get Recommendations"
          |
          v
Read current UI state (mode, selected user/title, filters, top_n)
          |
          v
Validate input for chosen mode
  - valid user ID?
  - matching movie title selected?
  - enough preference signal?
          |
          +---------------------------+
          |                           |
          v                           v
   Invalid / weak input         Valid input
   (show message + fallback)         |
          |                          v
          |                Load cached artifacts/resources
          |                (metadata, popularity tables,
          |                 genre features, CF similarity,
          |                 selected params)
          |                          |
          |                          v
          |                Build request context
          |                (seen items, seed movies, filters,
          |                 mode flags, candidate limits)
          |                          |
          |                          v
          |                Generate candidates by mode
          |                - hybrid personalized
          |                - similar-movie content path
          |                - cold-start genre/global fallback
          |                          |
          |                          v
          |                Rank + explain + filter + dedupe
          |                          |
          |                          v
          +-----------------------> Render Top-N cards
                                     + status message
                                     + explanation details
```

This diagram shows what happens after the user clicks the recommendation button and where the app should switch between validation, cached loading, ranking, and rendering. The user interaction is simple (select inputs and click), but behind the scenes the app routes the request to the right recommender path and fallback behavior; on screen, the result is a Top-N list plus context messages that explain which mode ran and whether fallback logic was used.

### Inference Pipeline Steps and Suggested Function Contracts

The implementation should use a small set of functions with clear inputs and outputs. Even if the final code is written in one Streamlit file at first, separating these responsibilities will reduce debugging time and make later refactoring into `src/serving/` easier.

| Pipeline stage | Suggested function (example name) | Inputs | Outputs |
| --- | --- | --- | --- |
| Artifact loading | `load_recommender_artifacts()` | `artifacts_dir` | Metadata tables, matrices, params, lookup maps |
| UI request parsing | `parse_ui_request()` | Streamlit widget values | Normalized request dict (mode, seeds, filters, top_n) |
| Validation | `validate_request()` | Request dict + metadata lookups | Validated request + warnings / fallback flags |
| Context building | `build_request_context()` | Validated request + user/profile artifacts | User seen set, seed items, preferred genres, filter context |
| Candidate generation | `generate_candidates()` | Context + artifacts | Candidate list(s) with raw scores and source labels |
| Ranking and explanation | `rank_and_explain()` | Candidates + params + metadata | Ranked recommendations with reason text |
| Post-filter and render prep | `prepare_display_rows()` | Ranked recs + metadata + filters | Display rows for cards + UI messages |

This table shows the recommended handoff between backend stages inside the Streamlit app. It matters for practical deployment because it defines where to place validation, fallback, and explanation logic, which makes both testing and user-visible behavior more predictable.

**Simple explanation:** a recommendation request should behave like a short checklist. The app reads what the user selected, checks that it is usable, runs the right recommender path, and then turns the result into readable cards with reasons.

## 5. Caching and Performance Strategy (Streamlit Runtime)

The caching strategy should follow the project rule from `AGENTS.md`: heavy preprocessing stays offline, while the Streamlit app loads precomputed artifacts and runs inference. In practice, this means the app should cache large immutable resources (matrices, metadata tables, lookup maps) at process scope and avoid recomputing search indexes or repeated parsing work on every widget change.

The most important caching boundary is between **artifact loading** and **request-time ranking**. Artifact loading is expensive but mostly static, so it should use `st.cache_resource`. Request-time ranking is fast enough for interactive use when candidate generation is bounded, so it should usually run directly, with small helper caches only where reuse is obvious and bounded.

A practical v1 caching plan is:
- Use `st.cache_resource` for heavy immutable artifacts and service objects (metadata table in memory, CF similarity matrix, genre feature matrix, parameter config, compiled lookup maps).
- Use `st.cache_data` for deterministic derived tables and search helpers (normalized title search index, prefiltered genre lists, optional poster lookup joins).
- Use `st.session_state` for user interaction state (selected mode, chosen movies, filters, last response, and UI messages).
- Avoid unbounded global caching of full user-specific recommendation results, because this can grow memory usage quickly and becomes harder to invalidate when artifacts change.

The main expected bottlenecks in v1 are startup artifact loading, title search normalization if rebuilt frequently, and ranking/filtering work when candidate sets are too large. Another major bottleneck appears if the app loads `ml-25m/ratings.csv` directly to support returning users; this is exactly why the K-Phase recommends a compact user-profile artifact.

For performance monitoring, the app should log (or display in a debug panel) coarse timings for: artifact load time, request validation, candidate generation, ranking/explanation, and total response time. This makes it easier to verify the Q-Phase responsiveness goal in the actual app, not only in AAA offline latency proxies.

### Caching and Runtime Boundaries (Plain-Language Deployment View)

```text
App startup (once per process)
  -> load artifacts and build lookup helpers (cached)

User interaction (many times)
  -> read widgets -> validate -> route -> rank -> render (not retraining, not rebuilding matrices)

When artifacts change
  -> clear cache / restart app -> reload new artifact set
```

This diagram shows the intended boundary between one-time startup work and repeated user interactions. It matters for user experience because the app should spend time loading heavy resources once, then respond quickly to clicks and filter changes instead of repeating the same expensive work on every rerun.

## 6. Error Handling and Edge Cases (User Trust and Stability)

A recommender app that fails silently will look unreliable even when the model is good. The K-Phase implementation should therefore treat error handling as part of the recommendation design, not as a later UI polish task.

The minimum v1 edge-case behavior should cover the cases required in `TASKS.MD` and map each one to a visible user message plus a deterministic backend response:

- **No matching movie search**: If title search returns no match, the app should show a clear message (for example, "No matching title found. Try a different spelling or a broader search.") and avoid running the similarity pipeline with an empty input.
- **Insufficient user preferences**: If new-user mode has too few liked titles or no usable genre signal, the app should say that it is using a cold-start fallback and then return genre/global popularity recommendations instead of an empty list.
- **Missing metadata**: If year, genre, or poster metadata is missing for a title, the app should still display the recommendation card with available fields and a neutral placeholder rather than dropping the recommendation silently.
- **Empty recommendation set**: If filtering removes all candidates, the app should explain that filters were too strict and either relax them in a controlled order or present a fallback list.

The backend should also surface internal validation failures in a user-safe way. For example, if a required artifact file is missing, the app should show a setup error message that names the missing file path and points the developer to the artifact export step, instead of raising a long stack trace in front of end users.

### Example User-Facing Message Strategy (Recommended Tone)

The app should use plain, non-technical messages that explain what happened and what the user can do next. This is especially important when fallback logic runs, because otherwise users may think the app ignored their input.

```text
Examples:
- "We could not find that movie title yet. Try another spelling or choose from the search suggestions."
- "You have not given enough preferences for personalized results yet, so we are showing popular picks in your selected genres."
- "No movies matched all current filters. We widened the year range to show a usable shortlist."
- "Setup issue: missing artifacts/cf_item_similarity.npz. Run the A-Phase export step before starting the app."
```

This message strategy shows the tone and level of detail the app should use when something goes wrong or when fallback behavior is triggered. It matters for trust because users are more likely to continue using the app when they understand why the result changed and what happened behind the scenes.
