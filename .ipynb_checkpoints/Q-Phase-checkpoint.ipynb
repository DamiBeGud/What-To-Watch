# Q-Phase: Defining the Recommendation Problem for the Movie Recommendation App

This notebook is the **Q (Question) phase** of the QUAAACK process for a **Movie Recommendation App** that will be delivered in **Streamlit**. The purpose of this phase is to define the recommendation problem clearly before any deep data exploration or model training begins.

In practical terms, this notebook answers a simple but important question: *what exactly are we trying to build, for whom, and how will we know if it is good enough to ship as a v1 demo?* That shared definition matters because a recommender can look impressive in a UI while still failing users if the goals, tradeoffs, and evaluation criteria are not explicit.

This Q-Phase therefore establishes the project context, the stakeholder expectations, the core recommendation questions, the v1 scope boundaries, the success metrics, and the main risks. It also previews how the later QUAAACK phases (U, AAA, C, K) will turn this problem definition into a working and explainable Streamlit application.


## 1. Project Overview (Plain Language)

The Movie Recommendation App is designed to help people discover movies they are likely to enjoy without spending too much time scrolling through a large catalog. A user should be able to enter the app, choose a simple starting point (for example a known profile or a movie they already like), and receive a ranked shortlist of recommendations with short explanations.

Recommendations matter because movie choice is a high-friction decision for many users. When a catalog is large, users often default to familiar titles or stop searching entirely. A good recommender reduces that friction, improves discovery, and makes the app feel useful quickly. In this project, recommendation quality is not only about getting "popular" movies onto the screen; it is about producing relevant suggestions that are understandable and defensible.

We use **Streamlit** because it supports fast iteration for interactive data products and is well suited for demos with mixed technical and non-technical audiences. It allows the project to focus on Python-based offline training and evaluation while serving precomputed artifacts in a responsive interface. This matches the accepted project decision to keep heavy training out of the request path and load artifacts at runtime instead (`DECISIONS.md`, ADR-002).


## 2. Problem Definition for Q-Phase

The v1 recommendation problem is framed as a **Top-N ranking task**: given a user context (history, liked movie(s), or filters), the system should produce a small ranked list of movies that are likely to be relevant and easy to act on. In the Streamlit app, this typically means a Top-10 output with title, year, genres, and a short explanation for each recommendation.

The project must support three common entry situations. First, a returning user may have rating history, which enables personalization. Second, a new user may have little or no history, which requires a cold-start strategy rather than collaborative personalization. Third, a user may simply want "movies similar to this one," which calls for item-to-item logic that is easy to explain.

This problem definition intentionally favors a practical balance of **quality, speed, explainability, and reproducibility** over maximum algorithmic complexity. That is consistent with the project architecture, the notebook workflow, and the need to present results clearly in a Streamlit demo.


## 3. Stakeholders and What Success Looks Like

### 3.1 End Users (Movie Watchers)
For end users, success is mostly about speed, relevance, and trust. They need recommendations that feel useful without requiring technical knowledge, and they need the app to respond quickly enough that the interaction feels smooth rather than delayed. They also need simple explanations, because a ranked list without context can feel random even when the ranking is technically strong.

Their main pain points are choice overload, repetitive popularity lists, and weak first-session experiences. If the app cannot help a new user or keeps showing the same obvious titles, it will not feel genuinely helpful. In this project, success for end users means they can get a plausible shortlist within seconds, understand why titles were suggested, and still receive helpful results even when they have no prior history in the system.

### 3.2 Product Owner / Project Team
The product owner and project team need a recommender that is both demonstrably useful and realistic to build in a v1 scope. Their priorities include a stable Streamlit demo, clear evidence that the recommendation logic outperforms simple baselines, and a system design that can evolve later without rewriting everything.

Their pain points are usually scope creep, unclear tradeoffs, and demos that look polished but do not hold up under questions about quality or method. For this stakeholder group, success means a presentable app, a clear evaluation story, and a documented rationale for why the chosen approach is appropriate for v1 rather than over-engineered.

### 3.3 Data / ML Developer
The Data/ML developer needs a problem definition that is precise enough to implement and evaluate safely. That includes leakage-aware split rules, clear ranking metrics, documented assumptions, and a practical serving target for Streamlit. They also need freedom to start with simple baselines and improve incrementally, because that makes debugging and communication much easier than jumping directly to a complex model.

Their pain points include ambiguous goals (for example, "make it smarter" without a metric), hidden leakage that inflates scores, and runtime/memory constraints from large files such as `ratings.csv` and especially `genome-scores.csv`. Success for the developer means the experiments are reproducible, the comparisons are fair, the artifacts are usable in the app, and the final recommendation behavior can be explained to non-technical reviewers.

### 3.4 Non-Technical Reviewers / Demo Audience
Non-technical reviewers need the notebook and the app to tell a clear story. They should be able to understand what the recommender is trying to do, why certain design choices were made, and what the main metrics mean in practical terms. They do not need deep implementation detail, but they do need honest explanations of limitations and tradeoffs.

Their pain points are jargon-heavy explanations and metric tables with no interpretation. Success for this group means they can follow the QUAAACK workflow, understand the app behavior in plain language, and see why the project chooses a practical and explainable v1 path.


## 4. Core Recommendation Questions (Q-Phase)

### 4.1 How do we recommend for users with history?
For users who have already rated movies, the recommender should use historical interactions to produce personalized results. In later phases, this will likely involve comparing collaborative and hybrid methods, but in Q-Phase the key decision is simpler: this path must exist and it must be evaluated with ranking metrics that reflect the Top-N experience in the app.

### 4.2 How do we recommend for new users (cold start)?
A new user has little or no interaction history, which means personalization signals may be weak or unavailable. The system therefore needs a deliberate fallback path, such as popularity-based recommendations refined by filters or basic content cues (for example genres), so the first app session still feels useful instead of empty.

### 4.3 How do we explain recommendations clearly?
Recommendation quality alone is not enough for this project. The app must also communicate *why* a title was suggested in simple terms, such as similarity to a liked movie, popularity within a selected genre, or relevance to the current filters. These explanations should support trust without pretending the recommendations are certain or perfect.

Together, these three questions define the minimum user experience that the v1 app must support: personalization for returning users, a fallback for newcomers, and transparent recommendation rationale for everyone.

## 6. Scope and Non-Scope for v1

The v1 scope is intentionally narrow enough to deliver a reliable and explainable Streamlit demo while still being meaningful as a recommender system. The focus is on offline-trained recommendation logic, ranking evaluation, and interactive inference in the UI. Heavy training, feature building, and model tuning happen outside the app runtime.

The following items are **in scope** for v1 because they directly support the demo goals and the QUAAACK workflow:
- personalized recommendations for users with history,
- cold-start recommendations for new users,
- similar-movie recommendations from a selected title,
- basic filtering and clear explanation text in Streamlit,
- offline evaluation with ranking metrics and coverage.

The following items are **out of scope** for v1 because they increase complexity without being necessary for the initial project objective:
- online learning or real-time retraining during app use,
- production-scale API deployment and autoscaling concerns,
- full commercial-grade metadata enrichment pipelines,
- highly complex models that reduce explainability without proven value for this use case.

Keeping this boundary explicit protects the project from feature creep and keeps the implementation aligned with the accepted decisions in `DECISIONS.md`, especially the separation between offline training and Streamlit inference.


## 7. Success Metrics (Business + ML + UX)

Success in this project must be measured in a way that reflects how users experience recommendations in the app. That is why the primary offline metrics are **ranking metrics** rather than only error metrics such as RMSE. The app presents a shortlist, so the quality question is whether the right movies appear near the top of that shortlist.

The required ranking metrics for the Top-10 recommendation task are:
- `Precision@10`
- `Recall@10`
- `NDCG@10`
- `HitRate@10`

In plain language, these metrics answer different parts of the same user-facing question: "Is the shortlist useful, and are the best suggestions shown near the top?" `Precision@10` focuses on how much of the list is relevant, `Recall@10` checks how much of the relevant content we recovered, `NDCG@10` rewards good ordering, and `HitRate@10` checks whether the user gets at least one strong recommendation in the list.

The project also tracks **catalog coverage**, because a recommender that always returns the same popular movies may look acceptable on some examples while failing the broader discovery goal. Coverage helps us monitor whether the system is surfacing a wider set of titles across users.

Finally, the Streamlit app needs a practical responsiveness target. For the v1 local demo experience, recommendation generation after a user selection should generally feel immediate and stay roughly under a few seconds, with a working target of about **under 2 seconds** for common interactions after artifacts are loaded. This is a product requirement as much as an engineering requirement because slow responses reduce trust and usability.


## 8. Risks and Constraints

The most important early risk is the **cold-start problem**. A recommender cannot personalize well for a user who has no history, and this can make the app feel broken on first use if a fallback path is not designed on purpose. That is why cold-start handling is treated as a core Q-Phase question rather than a later enhancement.

A second major risk is **popularity bias**. Recommenders can appear accurate by repeatedly recommending a small set of widely rated titles. While this may produce acceptable examples in a demo, it can weaken personalization and discovery. Coverage monitoring and comparison to genre-aware baselines will help identify this problem in later phases.

The project also faces **sparsity** in the user-item interaction matrix. Even with 25M ratings, the catalog is large and most users interact with only a small fraction of all movies. This affects collaborative methods in particular and makes robust baselines and leakage-safe evaluation design especially important.

There are additional practical constraints from the dataset itself. `ml-25m` is historical (activity ends in 2019), it reflects MovieLens users rather than all streaming audiences, and it provides less metadata than commercial systems. It also includes large files such as `genome-scores.csv`, which can slow notebook work and app runtime if used naively. These constraints do not prevent a strong educational recommender project, but they do define the limits of what can be claimed and what should be deferred beyond v1.


## 9. Data Used in Q-Phase (High-Level Only)

Q-Phase does not perform deep exploratory analysis. Instead, it documents the available `ml-25m` files at a high level and clarifies which ones are core for v1 and which ones are optional or advanced. This keeps the phase focused on problem definition while still grounding the design in the actual dataset contract (`DATA.md`).

At a high level, the dataset provides approximately **25,000,095 ratings**, **1,093,360 tag applications**, **62,423 movies**, and **162,541 users**, covering activity from **January 9, 1995 to November 21, 2019** (`ml-25m/README.txt`). The users are anonymized and no demographic information is included, which matters when interpreting what the recommender can and cannot personalize.

For v1, the essential data path uses `ratings.csv` for user-item interactions and `movies.csv` for title/genre metadata. The remaining files expand what is possible, especially for explainability and content enrichment, but they are not all necessary for an initial working app.

The file inventory for this project is listed explicitly below because Task 1 requires it:
- `ml-25m/ratings.csv` (core for personalized recommendations and offline evaluation)
- `ml-25m/movies.csv` (core for titles, genres, filters, and display)
- `ml-25m/tags.csv` (optional for v1; useful for content enrichment and explanations)
- `ml-25m/links.csv` (optional for v1; useful for external IDs/posters/metadata extension)
- `ml-25m/genome-scores.csv` (advanced optional; large file for rich content signals)
- `ml-25m/genome-tags.csv` (advanced optional; tag dictionary for genome features)
- `ml-25m/README.txt` (dataset documentation and limitations)

In later phases, `U-Phase.ipynb` will inspect quality, sparsity, and schema details, while `AAA-Phase` will decide which optional sources are worth the runtime and complexity tradeoff.


## 10. Optional Sanity Check (File Presence Only)

The next code cell is intentionally lightweight. It only checks whether the expected `ml-25m` files are present and prints a short summary paragraph from `ml-25m/README.txt`. This is not a data analysis step; it is a setup validation step.

This matters because it prevents avoidable path and file availability errors before the U-Phase begins. It also keeps Q-Phase aligned with the data contract rule that this phase should use only high-level dataset information, not heavy processing.


````python
from pathlib import Path

ml_dir = Path('ml-25m')
required_files = [
    'ratings.csv',
    'movies.csv',
    'tags.csv',
    'links.csv',
    'genome-scores.csv',
    'genome-tags.csv',
    'README.txt',
]

print('ml-25m file presence check (Q-Phase)\n')
for name in required_files:
    path = ml_dir / name
    status = 'FOUND' if path.exists() else 'MISSING'
    print(f'- {name}: {status}')

readme_path = ml_dir / 'README.txt'
if readme_path.exists():
    text = readme_path.read_text(encoding='utf-8', errors='replace')
    lines = text.splitlines()
    summary_lines = []
    in_summary = False
    for line in lines:
        stripped = line.strip()
        if not in_summary:
            if stripped.lower() == 'summary':
                in_summary = True
            continue
        if set(stripped) == {'='} or stripped == '':
            if summary_lines:
                break
            continue
        if stripped.lower() == 'usage license':
            break
        summary_lines.append(stripped)
    if summary_lines:
        print('\nREADME summary paragraph:')
        print(' '.join(summary_lines))
````

This roadmap diagram shows the project sequence after the Q-Phase establishes the problem definition. The arrows indicate progression, and each phase produces a different kind of decision or artifact rather than repeating the same work in a different notebook.

For non-technical readers, the key idea is simple: we first define the question, then verify the data can support it, then build and evaluate recommendation approaches, then choose what to ship, and finally document everything so the Streamlit app and future contributors can use the result reliably.

### What each next phase will do (short preview)
The **U-Phase** will focus on understanding the `ml-25m` dataset in detail, including schema checks, data quality, sparsity, long-tail effects, and the rationale for leakage-safe chronological splits. The **AAA-Phase** will build baselines and candidate recommenders, adapt features, and evaluate them with ranking metrics and explanation examples. The **C-Phase** will summarize tradeoffs and select the v1 approach to ship. The **K-Phase** will document the pipeline, the saved artifacts, and the app-facing recommendation logic for reuse and handoff.


## 12. Summary and Interpretation (End of Q-Phase)

This Q-Phase defines the Movie Recommendation App as a **practical Top-N recommendation problem** for a Streamlit interface, with explicit attention to explainability, responsiveness, and reproducible offline evaluation. The notebook clarifies who the app serves, what recommendation experiences must be supported, and which success measures will be used to judge the later modeling work.

Just as importantly, it sets realistic boundaries. The project is not trying to build a production-scale streaming platform in v1. It is building an explainable and credible recommender demo that handles both personalization and cold start, uses `ml-25m` responsibly, and can be defended to technical and non-technical audiences.

The next step is `U-Phase.ipynb`, where the project moves from problem framing to data understanding. That notebook will test whether the dataset quality, sparsity patterns, and feature availability support the design decisions made here, and it will establish the leakage-safe foundations needed for fair model evaluation in the AAA phase.
